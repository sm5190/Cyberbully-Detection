Clearing html marks (&amp)

```python
from bs4 import BeautifulSoup
```

```python
def html_char_removal(row):
    text = row['Text']
    text = BeautifulSoup(text).get_text()

    return text
```

```python
train_df['Text'] = train_df.apply(html_char_removal, axis=1)
```

```python
!pip install gensim
```

```python
from gensim.parsing.preprocessing import remove_stopwords
```

```python
def stopword_removal(row):
    text = row['Text']
    text = remove_stopwords(text)
    return text
```

```python
train_df['Text'] = train_df.apply(stopword_removal, axis=1)
```

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()
```

```python
!pip install twython 
```

```python
train_df['scores'] = train_df['Text'].apply(lambda Text: sid.polarity_scores(Text))

train_df.head()
```

```python
train_df['compound']  = train_df['scores'].apply(lambda score_dict: score_dict['compound'])

train_df.head()
```

```python
train_df['comp_score'] = train_df['compound'].apply(lambda c: 'neg' if c<=0  else 'pos')

train_df.head()
```
